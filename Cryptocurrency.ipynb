```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
df_bitdash=pd.merge(bitcoin_df,dash_df,on='Date',how="outer")
df_bitdash.head()
```

```python
# Rename columns so that they are differentiated
df_bitdash.rename(columns={'Open_x':'Open Bitcoin', 'High_x':'High Bitcoin', 'Low_x':'Low Bitcoin', 
                           'Close_x':'Close Bitcoin', 'Volume_x':'Volume Bitcoin',
       'Market Cap_x':'Market Cap Bitcoin', 'Open_y':'Open Dash', 'High_y':'High Dash', 'Low_y':'Low Dash',
                           'Close_y':'Close Dash', 'Volume_y':'Volume Dash','Market Cap_y':'Market Cap Dash',
       },inplace=True)
```

```python
df_bitdash.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
```

```python
# Collecting best open for Bitcoin and Dash
best_open_bit=df_bitdash["Open Bitcoin"].max()
best_open_dash=df_bitdash["Open Dash"].max()
# Collecting best close for Bitcoin and Dash
best_close_bit=df_bitdash["Close Bitcoin"].max()
best_close_dash=df_bitdash["Close Dash"].max()
# Collecting the total volume for Bitcoin and Dash
best_volume_bit=df_bitdash["Volume Bitcoin"].sum()
best_volume_dash=df_bitdash["Volume Dash"].sum()
```

```python
# Creating a summary DataFrame using above values
#data_BD=({"Highest_Open_Bitcoin":best_open,"Highest_Close_Bitcoin":best_close,"Highest_Volume_Bitcoin":best_volume})
df_final=pd.DataFrame({"Highest Open Bitcoin":[best_open_bit],"Highest Close Bitcoin":[best_close_bit],
                       "Highest Open Dash":[best_open_dash],"Highest Close Dash":[best_close_dash],
                       "Highest Volume Bitcoin":[best_volume_bit],"Highest Volume Dash":[best_volume_dash]})
df_final
```
