```python
import pandas as pd
```

```python
# The path to our CSV file
file="Resources/KickstarterData.csv"
# Read our Kickstarter data into pandas
df=pd.read_csv(file)
df.head()
```

```python
# Get a list of all of our columns for easy reference
df.columns
```

```python
# Extract "name", "goal", "pledged", "state", "country", "staff_pick",
# "backers_count", and "spotlight"
df_kick=df.loc[:,(["name", "goal", "pledged", "state", "country", "staff_pick","backers_count","spotlight"])]
df_kick.head()
```

```python
# Remove projects that made no money at all
df_kick1=df_kick.loc[(df_kick["pledged"]>0)]
df_kick1.head()
```

```python
# Collect only those projects that were hosted in the US
df_kick1.loc[(df_kick1["country"]=="US")]
# Create a list of the columns
columns=["name", "goal", "pledged", "state", "country", "staff_pick","backers_count","spotlight"]
# Create a new df for "US" with the columns above. 
df_kick2=df_kick1.loc[(df_kick1["country"]=="US")]
df_kick2.head()
```

```python
# Create a new column that finds the average amount pledged to a project
df_kick2["average donation"]=df_kick2["pledged"]/df_kick2["backers_count"]
df_kick2.head()
```

```python
# First convert "average_donation", "goal", and "pledged" columns to float
# Then Format to go to two decimal places, include a dollar sign, and use comma notation

df_kick2["average donation"]=df_kick2["average donation"].astype(float).map("${:,.2F}".format)
df_kick2["goal"]=df_kick2["goal"].astype(float).map("${:,.2F}".format)
df_kick2["pledged"]=df_kick2["pledged"].astype(float).map("${:,.2F}".format)
df_kick2.head()
```

```python
# Calculate the total number of backers for all US projects
df_kick2["backers_count"].sum()
```

```python
# Calculate the average number of backers for all US projects
df_kick2["backers_count"].mean()
```

```python
# Collect only those US campaigns that have been picked as a "Staff Pick"
df_kick3=df_kick2.loc[df_kick2["staff_pick"]==True]
df_kick3.head()
```

```python
# Group by the state of the campaigns and see if staff picks matter (Seems to matter quite a bit)
df_kick4=df_kick3.groupby("state")
df_kick4.head()
```
