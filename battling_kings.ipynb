```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
file="Resources/got.csv"
df=pd.read_csv(file)
df.head()
```

```python
# Get attacker and defender data
attacker = df["attacker_king"].value_counts()
defender = df["defender_king"].value_counts()
```

```python
# Get total battle data
battle_df=attacker.add(defender,fill_value=0)
battle_df
```

```python
x_axis = np.arange(len(battle_df))
tick_locations = [value for value in x_axis]
```

```python
battle_df.plot(kind="bar",facecolor="red")
#plt.xticks(tick_locations, battle_df, rotation="vertical")
plt.title("The Bloodthirst of Kings")
plt.xlabel("King")
plt.ylabel("Number of Battle Participated in")
plt.show()
```
