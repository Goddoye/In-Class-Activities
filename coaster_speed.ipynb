```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
Danger_Drop= [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
RailGun= [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
sec= np.arange(0,121,10)
```

```python
plt.plot(sec, Danger_Drop,color="red",label="Danger Drop")
plt.plot(sec, RailGun,color="green",label="Rail Gun")
plt.grid()
plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
plt.xlim(0,120)
plt.ylim(0,100)
plt.legend(loc="upper right")
```
