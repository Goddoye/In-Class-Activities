```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Fahrenheit= [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
months= np.arange(1,13,1)
```

```python
plt.plot(months,Fahrenheit,marker="+",color="blue",label="Fahrenheit")
```

```python
celsius= [(x-32) * 0.56 for x in Fahrenheit]
```

```python
plt.plot(months,celsius,marker="s",color="red",label="Celsius")
```

```python
plt.xlabel("Months")
plt.ylabel("Degree")
```

```python
plt.legend(loc="upper right")
```
