```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Fahrenheit= [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
months= np.arange(1,13,1)
```

```python
months
```

```python
plt.plot(months,Fahrenheit)
```

```python
celsius= [(x-32) * 0.56 for x in Fahrenheit]
```

```python
celsius
```

```python
plt.plot(months,celsius)
```

```python
plt.plot(months,Fahrenheit,"m--")
plt.plot(months,celsius,'k:')
```

```python
plt.xlabel("Months")
plt.ylabel("Temprature: F(Mangenta),C(Black)")
plt.plot(months,Fahrenheit,"m--")
plt.plot(months,celsius,'k:')
```
